```python
# 🔐 ОСНОВНАЯ КОНФИГУРАЦИЯ
import os
import json
import asyncio
from datetime import datetime
import nest_asyncio
import requests
import base64
from sentence_transformers import SentenceTransformer, util
import torch

# Применяем для работы асинхронности в Colab
nest_asyncio.apply()

class Config:
    # Настройки GigaChat
    GIGACHAT_CLIENT_ID = 
    GIGACHAT_CLIENT_SECRET = 
    GIGACHAT_SCOPE = "GIGACHAT_API_PERS"
    GIGACHAT_AUTH_URL = "https://ngw.devices.sberbank.ru:9443/api/v2/oauth"
    GIGACHAT_API_URL = "https://gigachat.devices.sberbank.ru/api/v1"

    # База вакансий
    VACANCIES = {
        "Python Разработчик": {
            "required_skills": ["Python", "Django", "PostgreSQL", "Docker", "Git", "Linux"],
            "description": "Разработка веб-приложений на Python",
            "min_salary": 120000,
            "max_salary": 250000
        },
        "Data Scientist": {
            "required_skills": ["Python", "ML", "Pandas", "SQL", "Statistics", "NumPy"],
            "description": "Анализ данных и машинное обучение",
            "min_salary": 140000,
            "max_salary": 300000
        },
        "DevOps Engineer": {
            "required_skills": ["Docker", "Kubernetes", "AWS", "Linux", "CI/CD", "Terraform"],
            "description": "Развертывание и поддержка инфраструктуры",
            "min_salary": 150000,
            "max_salary": 320000
        },
        "Frontend Разработчик": {
            "required_skills": ["JavaScript", "React", "TypeScript", "HTML", "CSS", "Vue"],
            "description": "Разработка пользовательских интерфейсов",
            "min_salary": 100000,
            "max_salary": 220000
        }
    }

    # Порог соответствия
    MATCH_THRESHOLD = 70
    BOT_TOKEN = 
    CHANNEL_ID = 
    ADMIN_USERNAME = 

print("✅ Конфигурация загружена!")
```

```python
# 🧠 GIGACHAT КЛИЕНТ ДЛЯ АНАЛИЗА РЕЗЮМЕ
class GigaChatClient:
    def __init__(self):
        self.config = Config()
        self.access_token = None
        self._get_access_token()

    def _get_basic_auth(self):
        credentials = f"{self.config.GIGACHAT_CLIENT_ID}:{self.config.GIGACHAT_CLIENT_SECRET}"
        return base64.b64encode(credentials.encode()).decode()

    def _get_access_token(self):
        headers = {
            'Content-Type': 'application/x-www-form-urlencoded',
            'Authorization': f'Basic {self._get_basic_auth()}',
            'RqUID': self.config.GIGACHAT_CLIENT_ID,
        }
        data = {'scope': self.config.GIGACHAT_SCOPE}

        try:
            response = requests.post(self.config.GIGACHAT_AUTH_URL, headers=headers, data=data, verify=False)
            if response.status_code == 200:
                self.access_token = response.json()['access_token']
                print("✅ GigaChat токен получен успешно")
            else:
                print(f"❌ Ошибка получения токена: {response.status_code}")
        except Exception as e:
            print(f"❌ Ошибка получения токена: {e}")

    def extract_skills(self, text):
        """Извлечение навыков из текста резюме"""
        prompt = f"""
        Извлеки технические навыки и компетенции из текста резюме.
        Верни ТОЛЬКО JSON: {{"skills": ["skill1", "skill2", "skill3"]}}

        Текст резюме: {text[:3000]}
        """

        messages = [
            {"role": "system", "content": "Ты HR-эксперт. Извлекаешь навыки из резюме. Отвечай только JSON."},
            {"role": "user", "content": prompt}
        ]

        return self._get_chat_response(messages)

    def analyze_resume_match(self, resume_skills, vacancy_name, vacancy_skills):
        """Анализ соответствия резюме вакансии"""
        prompt = f"""
        Проанализируй соответствие кандидата вакансии и рассчитай процент соответствия резюме.

        ВАКАНСИЯ: {vacancy_name}
        ТРЕБУЕМЫЕ НАВЫКИ: {', '.join(vacancy_skills)}

        НАВЫКИ КАНДИДАТА: {', '.join(resume_skills)}

        Верни ТОЛЬКО JSON:
        {{
            "match_percentage": 85,
            "matched_skills": ["Python", "Django"],
            "missing_skills": ["Docker", "Kubernetes"],
            "recommendation": "Рекомендован к собеседованию"
        }}
        """

        messages = [
            {"role": "system", "content": "Ты Senior HR-аналитик. Анализируешь соответствие кандидатов."},
            {"role": "user", "content": prompt}
        ]

        return self._get_chat_response(messages)

    def _get_chat_response(self, messages):
        if not self.access_token:
            print("❌ Нет access token")
            return None

        headers = {
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {self.access_token}',
        }

        payload = {
            'model': 'GigaChat',
            'messages': messages,
            'temperature': 0.3
        }

        try:
            response = requests.post(
                f"{self.config.GIGACHAT_API_URL}/chat/completions",
                headers=headers,
                json=payload,
                verify=False,
                timeout=30
            )

            if response.status_code == 200:
                return response.json()['choices'][0]['message']['content']
            else:
                print(f"❌ Ошибка API: {response.status_code}")
        except Exception as e:
            print(f"❌ Ошибка GigaChat: {e}")

        return None

print("✅ GigaChat клиент готов!")
```

```python
# 📊 АНАЛИЗАТОР РЕЗЮМЕ
class ResumeAnalyzer:
    def __init__(self):
        self.giga_client = GigaChatClient()
        self.config = Config()
        self.skill_model = SentenceTransformer('all-MiniLM-L6-v2')

    def analyze_resume_for_vacancies(self, resume_text):
        """Анализ резюме для всех вакансий"""
        print("🔍 Анализируем резюме...")

        # Извлекаем навыки из резюме
        skills_response = self.giga_client.extract_skills(resume_text)
        resume_skills = self._parse_skills_from_response(skills_response)

        if not resume_skills:
            # Резервный метод извлечения навыков
            resume_skills = self._fallback_skill_extraction(resume_text)

        print(f"✅ Найдены навыки: {resume_skills}")

        # Анализируем соответствие каждой вакансии
        results = {}
        for vacancy_name, vacancy_data in self.config.VACANCIES.items():
            print(f"📋 Проверяем вакансию: {vacancy_name}")

            match_result = self.giga_client.analyze_resume_match(
                resume_skills,
                vacancy_name,
                vacancy_data["required_skills"]
            )

            # Парсим результат
            vacancy_match = self._parse_match_result(match_result, vacancy_data)
            results[vacancy_name] = vacancy_match

            print(f"   📊 Соответствие: {vacancy_match['match_percentage']}%")

        return results, resume_skills

    def _parse_skills_from_response(self, response):
        """Парсинг навыков из ответа GigaChat"""
        if not response:
            return []

        try:
            import re
            json_match = re.search(r'\{.*\}', response, re.DOTALL)
            if json_match:
                data = json.loads(json_match.group())
                return data.get('skills', [])
        except Exception as e:
            print(f"⚠️ Ошибка парсинга навыков: {e}")

        return []

    def _parse_match_result(self, response, vacancy_data):
        """Парсинг результата анализа соответствия"""
        default_result = {
            "match_percentage": 0,
            "matched_skills": [],
            "missing_skills": vacancy_data["required_skills"],
            "recommendation": "Не соответствует",
            "vacancy_data": vacancy_data
        }

        if not response:
            return default_result

        try:
            import re
            json_match = re.search(r'\{.*\}', response, re.DOTALL)
            if json_match:
                data = json.loads(json_match.group())
                data["vacancy_data"] = vacancy_data
                return data
        except Exception as e:
            print(f"❌ Ошибка парсинга результата: {e}")

        return default_result

    def _fallback_skill_extraction(self, text):
        """Резервный метод извлечения навыков"""
        common_skills = [
            "Python", "Java", "JavaScript", "SQL", "Linux", "Docker", "Kubernetes",
            "AWS", "Git", "React", "PostgreSQL", "MongoDB", "Django", "Flask",
            "Machine Learning", "Pandas", "NumPy", "HTML", "CSS", "TypeScript",
            "C++", "C#", "PHP", "Ruby", "Go", "Rust", "Swift", "Kotlin"
        ]

        found_skills = []
        text_lower = text.lower()

        for skill in common_skills:
            if skill.lower() in text_lower:
                found_skills.append(skill)

        return found_skills

print("✅ Анализатор резюме готов!")
```

```python
# 🤖 TELEGRAM БОТ С ОТПРАВКОЙ В КАНАЛ
from telegram import Update, ReplyKeyboardMarkup, KeyboardButton
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes

# Импорты для работы с файлами
try:
    import pdfplumber
    import docx
    HAS_DOCUMENT_SUPPORT = True
except ImportError as e:
    print(f"⚠️ Предупреждение: {e}")
    HAS_DOCUMENT_SUPPORT = False

class HRTelegramBot:
    def __init__(self):
        self.config = Config()
        self.analyzer = ResumeAnalyzer()
        self.user_data = {}

    def get_main_keyboard(self):
        """Клавиатура главного меню"""
        keyboard = [
            [KeyboardButton("📄 Загрузить резюме"), KeyboardButton("📝 Создать резюме")],
            [KeyboardButton("🔍 Найти вакансии"), KeyboardButton("ℹ️ Помощь")]
        ]
        return ReplyKeyboardMarkup(keyboard, resize_keyboard=True)

    async def send_to_channel(self, context: ContextTypes.DEFAULT_TYPE, message: str):
        """Функция для отправки сообщений в канал"""
        try:
            await context.bot.send_message(
                chat_id=self.config.CHANNEL_ID,
                text=message,
                parse_mode='HTML'
            )
            print(f"✅ Сообщение отправлено в канал {self.config.CHANNEL_ID}")
            return True
        except Exception as e:
            print(f"❌ Ошибка отправки в канал: {e}")
            return False

    async def notify_new_user(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Уведомление о новом пользователе в канал"""
        user = update.effective_user

        welcome_message = f"""
<b>🆕 Новый пользователь в HR-боте!</b>

👤 <b>Информация:</b>
Имя: {user.first_name}
Фамилия: {user.last_name or 'не указана'}
Username: @{user.username or 'не указан'}
ID: {user.id}

⏰ Время: {update.message.date}
"""
        await self.send_to_channel(context, welcome_message)

    async def notify_resume_analysis(self, context: ContextTypes.DEFAULT_TYPE, user, results, skills):
        """Уведомление о результатах анализа резюме в канал"""
        # Находим подходящие вакансии
        suitable_vacancies = {
            name: data for name, data in results.items()
            if data['match_percentage'] >= self.config.MATCH_THRESHOLD
        }

        if suitable_vacancies:
            message = f"""
<b>🎯 НОВЫЙ КАНДИДАТ ДЛЯ ВАКАНСИЙ!</b>

👤 <b>Кандидат:</b>
Имя: {user.first_name}
Username: @{user.username or 'не указан'}
ID: {user.id}

🛠️ <b>Навыки кандидата:</b>
{', '.join(skills[:10])}

✅ <b>Подходящие вакансии:</b>
"""
            for vacancy_name, data in suitable_vacancies.items():
                message += f"\n🏢 <b>{vacancy_name}</b> - {data['match_percentage']}%"
                message += f"\n💼 {data['vacancy_data']['description']}"
                message += f"\n💰 {data['vacancy_data']['min_salary']:,} - {data['vacancy_data']['max_salary']:,} руб."
                if data['matched_skills']:
                    message += f"\n✅ Совпавшие навыки: {', '.join(data['matched_skills'][:3])}\n"

            message += f"\n⏰ Анализ проведен: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"

        else:
            # Лучший результат даже если не подходит
            if results:
                best_match_name, best_match_data = max(results.items(), key=lambda x: x[1]['match_percentage'])
                message = f"""
<b>📊 КАНДИДАТ НЕ ПРОШЕЛ ОТБОР</b>

👤 <b>Кандидат:</b>
Имя: {user.first_name}
Username: @{user.username or 'не указан'}

🛠️ <b>Навыки:</b>
{', '.join(skills[:8])}

📈 <b>Лучшее соответствие:</b>
{best_match_name} - {best_match_data['match_percentage']}%

💡 <b>Рекомендация:</b>
Требуется дополнительное обучение

⏰ Время анализа: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
"""
            else:
                message = f"""
<b>📊 ОШИБКА АНАЛИЗА</b>

👤 <b>Кандидат:</b>
Имя: {user.first_name}

❌ Не удалось проанализировать резюме
"""

        await self.send_to_channel(context, message)

    async def start(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Команда начала работы"""
        user = update.effective_user

        welcome_text = f"""
👋 Добро пожаловать, {user.first_name}!

Я - HR-бот с искусственным интеллектом. Я помогу вам:

📄 Анализ резюме - загрузите ваше резюме
📝 Создание резюме - помогу составить профессиональное резюме
🔍 Поиск вакансий - найду подходящие вакансии по вашему опыту

Выберите действие ниже:
"""

        await update.message.reply_text(welcome_text, reply_markup=self.get_main_keyboard())

        # Отправляем уведомление в канал о новом пользователе
        await self.notify_new_user(update, context)

        print(f"👤 Новый пользователь: {user.first_name} (@{user.username})")

    async def handle_upload_resume(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Обработка загрузки резюме"""
        if not HAS_DOCUMENT_SUPPORT:
            await update.message.reply_text(
                "❌ Поддержка документов временно недоступна\n\n"
                "Используйте опцию '📝 Создать резюме' для создания резюме в чате."
            )
            return

        instruction = """
📄 Загрузка резюме

Пожалуйста, отправьте ваше резюме в одном из форматов:
• PDF
• DOC/DOCX
• TXT

Я проанализирую его и найду подходящие вакансии!
"""
        await update.message.reply_text(instruction)

    async def handle_create_resume(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Создание резюме через диалог"""
        questions = [
            "Как вас зовут (ФИО)?",
            "Какая у вас профессия/специальность?",
            "Сколько лет опыта работы?",
            "Какие ключевые навыки и технологии вы знаете?",
            "Ваш уровень образования?",
            "Какие проекты или достижения хотите отметить?",
            "Ваши контактные данные (email/телефон)?"
        ]

        # Сохраняем вопросы в контексте пользователя
        user_id = update.effective_user.id
        self.user_data[user_id] = {
            'resume_questions': questions,
            'current_question': 0,
            'resume_answers': [],
            'resume_text': ""
        }

        await update.message.reply_text(
            "📝 Создание резюме\n\n"
            "Отлично! Давайте создадим ваше профессиональное резюме.\n"
            "Я задам вам несколько вопросов.\n\n"
            f"❓ Вопрос 1/{len(questions)}:\n{questions[0]}"
        )

    async def handle_find_vacancies(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Поиск вакансий без резюме"""
        instruction = """
🔍 Поиск вакансий

Для поиска подходящих вакансий мне нужно ваше резюме.

Вы можете:
1. 📄 Загрузить готовое резюме
2. 📝 Создать резюме прямо здесь

После анализа я покажу вакансии, где ваше соответствие более 70%!
"""
        await update.message.reply_text(instruction, reply_markup=self.get_main_keyboard())

    async def handle_document(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Обработка загруженных документов"""
        if not HAS_DOCUMENT_SUPPORT:
            await update.message.reply_text(
                "❌ Поддержка документов временно недоступна\n\n"
                "Используйте опцию '📝 Создать резюме' для создания резюме в чате."
            )
            return

        document = update.message.document
        user = update.effective_user

        # Проверяем формат файла
        allowed_extensions = ['.pdf', '.doc', '.docx', '.txt']
        file_extension = f".{document.file_name.split('.')[-1].lower()}"

        if file_extension not in allowed_extensions:
            await update.message.reply_text(
                "❌ Неверный формат файла!\n\n"
                "Поддерживаемые форматы: PDF, DOC, DOCX, TXT\n"
                f"Ваш файл: {document.file_name}"
            )
            return

        # Уведомляем о начале обработки
        processing_msg = await update.message.reply_text(
            "🔄 Обрабатываю ваше резюме...\n\n"
            "Анализирую навыки и ищу подходящие вакансии.\n"
            "Это займет несколько секунд..."
        )

        try:
            # Скачиваем файл
            file = await document.get_file()
            file_path = f"temp_{user.id}_{document.file_name}"
            await file.download_to_drive(file_path)

            # Извлекаем текст
            resume_text = self.extract_text_from_file(file_path)

            if not resume_text:
                await update.message.reply_text(
                    "❌ Не удалось прочитать файл\n\n"
                    "Попробуйте другой формат или создайте резюме в чате."
                )
                return

            # Анализируем резюме
            results, skills = self.analyzer.analyze_resume_for_vacancies(resume_text)

            # Формируем и отправляем результаты
            await self.send_analysis_results(update, context, results, skills, user)

            # Удаляем временный файл
            import os
            if os.path.exists(file_path):
                os.remove(file_path)

        except Exception as e:
            await update.message.reply_text(
                f"❌ Ошибка обработки: {str(e)}\n\n"
                "Попробуйте еще раз или создайте резюме в чате."
            )

    def extract_text_from_file(self, file_path):
        """Извлечение текста из файла"""
        text = ""
        try:
            if file_path.endswith('.pdf'):
                with pdfplumber.open(file_path) as pdf:
                    for page in pdf.pages:
                        if page.extract_text():
                            text += page.extract_text() + "\n"
            elif file_path.endswith('.docx'):
                doc = docx.Document(file_path)
                text = "\n".join([paragraph.text for paragraph in doc.paragraphs])
            elif file_path.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8') as f:
                    text = f.read()
        except Exception as e:
            print(f"❌ Ошибка чтения файла: {e}")
        return text

    async def send_analysis_results(self, update: Update, context: ContextTypes.DEFAULT_TYPE, results, skills, user):
        """Отправка результатов анализа"""
        # Находим подходящие вакансии (match > 70%)
        suitable_vacancies = {
            name: data for name, data in results.items()
            if data['match_percentage'] >= self.config.MATCH_THRESHOLD
        }

        # Формируем сообщение для пользователя
        if suitable_vacancies:
            message = "🎉 Отличные новости!\n\n"
            message += "✅ Вы подходите на следующие вакансии:\n\n"

            for vacancy_name, data in suitable_vacancies.items():
                message += f"🏢 {vacancy_name}\n"
                message += f"📊 Соответствие: {data['match_percentage']}%\n"
                message += f"💼 {data['vacancy_data']['description']}\n"
                message += f"💰 Зарплата: {data['vacancy_data']['min_salary']:,} - {data['vacancy_data']['max_salary']:,} руб.\n"

                if data['matched_skills']:
                    message += f"✅ Ваши навыки: {', '.join(data['matched_skills'][:5])}\n"

                message += "\n"

            message += "📞 Менеджер компании свяжется с вами в ближайшее время!\n\n"
            message += "Хотите найти другие вакансии? Загрузите обновленное резюме!"

        else:
            message = "📊 Результаты анализа:\n\n"
            message += "К сожалению, по текущему резюме нет вакансий с высоким соответствием.\n\n"

            # Показываем лучший результат
            if results:
                best_match = max(results.items(), key=lambda x: x[1]['match_percentage'])
                message += f"🧭 Самый близкий вариант: {best_match[0]}\n"
                message += f"📊 Соответствие: {best_match[1]['match_percentage']}%\n\n"

            message += "💡 Рекомендации:\n"
            message += "• Обновите резюме, добавив больше навыков\n"
            message += "• Рассмотрите смежные направления\n"
            message += "• Пройдите дополнительные курсы\n\n"
            message += "Попробуйте создать резюме заново с моей помощью!"

        await update.message.reply_text(message, reply_markup=self.get_main_keyboard())

        # 🔥 ОТПРАВЛЯЕМ УВЕДОМЛЕНИЕ В КАНАЛ
        try:
            await self.notify_resume_analysis(context, user, results, skills)
        except Exception as e:
            print(f"⚠️ Не удалось отправить в канал: {e}")

        # Логируем результат
        print(f"📊 Анализ для {user.first_name}: {len(suitable_vacancies)} подходящих вакансий")

    async def handle_text_message(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """Обработка текстовых сообщений"""
        text = update.message.text
        user_id = update.effective_user.id

        # Проверяем, находится ли пользователь в процессе создания резюме
        if user_id in self.user_data and 'resume_questions' in self.user_data[user_id]:
            await self.handle_resume_creation(update, user_id, context)
            return

        # Обработка команд из кнопок
        if text == "📄 Загрузить резюме":
            await self.handle_upload_resume(update, context)
        elif text == "📝 Создать резюме":
            await self.handle_create_resume(update, context)
        elif text == "🔍 Найти вакансии":
            await self.handle_find_vacancies(update, context)
        elif text == "ℹ️ Помощь":
            await self.send_help(update)
        else:
            await update.message.reply_text(
                "Я не понял ваш запрос. Используйте кнопки ниже для навигации.",
                reply_markup=self.get_main_keyboard()
            )

    async def handle_resume_creation(self, update: Update, user_id, context: ContextTypes.DEFAULT_TYPE):
        """Обработка процесса создания резюме"""
        user_data = self.user_data[user_id]
        questions = user_data['resume_questions']
        current_index = user_data['current_question']
        answers = user_data['resume_answers']

        # Сохраняем текущий ответ
        answers.append(update.message.text)

        # Переходим к следующему вопросу
        next_index = current_index + 1

        if next_index < len(questions):
            user_data['current_question'] = next_index
            await update.message.reply_text(
                f"✅ Ответ сохранен!\n\n"
                f"❓ Вопрос {next_index + 1}/{len(questions)}:\n{questions[next_index]}"
            )
        else:
            # Все вопросы отвечены, создаем резюме
            await self.create_final_resume(update, user_id, answers, context)

    async def create_final_resume(self, update: Update, user_id, answers, context: ContextTypes.DEFAULT_TYPE):
        """Создание финального резюме и анализ"""
        # Формируем текст резюме
        resume_text = self.format_resume_text(answers)

        # Сохраняем текст резюме
        self.user_data[user_id]['resume_text'] = resume_text

        # Показываем резюме пользователю
        await update.message.reply_text(
            "📋 Ваше резюме готово!\n\n"
            f"{resume_text}\n\n"
            "🔄 Анализирую вакансии..."
        )

        # Анализируем резюме
        results, skills = self.analyzer.analyze_resume_for_vacancies(resume_text)

        # Отправляем результаты
        await self.send_analysis_results(update, context, results, skills, update.effective_user)

        # Очищаем данные пользователя
        del self.user_data[user_id]

    def format_resume_text(self, answers):
        """Форматирование текста резюме из ответов"""
        name, profession, experience, skills, education, projects, contacts = answers

        resume_text = f"""
👤 ФИО: {name}
💼 Профессия: {profession}
📅 Опыт работы: {experience}
🛠️ Навыки: {skills}
🎓 Образование: {education}
🚀 Проекты и достижения: {projects}
📞 Контакты: {contacts}
"""
        return resume_text

    async def send_help(self, update: Update):
        """Отправка справки"""
        help_text = """
ℹ️ Помощь по боту

🤖 Что я умею:
• Анализировать ваше резюме с помощью ИИ
• Находить подходящие вакансии
• Помогать создавать профессиональные резюме
• Оценивать соответствие требованиям работодателей

📋 Как пользоваться:
1. 📄 Загрузить резюме - отправьте файл (PDF, DOC, DOCX, TXT)
2. 📝 Создать резюме - ответьте на вопросы для составления резюме
3. 🔍 Найти вакансии - узнайте, куда вы подходите

🎯 Критерии отбора:
• Вакансии с соответствием >70% считаются подходящими
• Менеджер свяжется с вами по подходящим вакансиям

По всем вопросам: @hr_manager
"""
        await update.message.reply_text(help_text)

    def run_bot(self):
        """Запуск бота"""
        app = Application.builder().token(self.config.BOT_TOKEN).build()

        # Обработчики команд
        app.add_handler(CommandHandler("start", self.start))
        app.add_handler(CommandHandler("help", self.send_help))

        # Обработчики сообщений
        app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, self.handle_text_message))

        if HAS_DOCUMENT_SUPPORT:
            app.add_handler(MessageHandler(filters.Document.ALL, self.handle_document))

        print("🤖 HR Bot запущен...")
        print("🔄 Ожидаем сообщения...")

        app.run_polling()

print("✅ Telegram бот готов!")
```
