```python
# üß† GIGACHAT –ö–õ–ò–ï–ù–¢ –î–õ–Ø –ê–ù–ê–õ–ò–ó–ê –†–ï–ó–Æ–ú–ï
class GigaChatClient:
    def __init__(self):
        self.config = Config()
        self.access_token = None
        self._get_access_token()

    def _get_basic_auth(self):
        credentials = f"{self.config.GIGACHAT_CLIENT_ID}:{self.config.GIGACHAT_CLIENT_SECRET}"
        return base64.b64encode(credentials.encode()).decode()

    def _get_access_token(self):
        headers = {
            'Content-Type': 'application/x-www-form-urlencoded',
            'Authorization': f'Basic {self._get_basic_auth()}',
            'RqUID': self.config.GIGACHAT_CLIENT_ID,
        }
        data = {'scope': self.config.GIGACHAT_SCOPE}

        try:
            response = requests.post(self.config.GIGACHAT_AUTH_URL, headers=headers, data=data, verify=False)
            if response.status_code == 200:
                self.access_token = response.json()['access_token']
                print("‚úÖ GigaChat —Ç–æ–∫–µ–Ω –ø–æ–ª—É—á–µ–Ω —É—Å–ø–µ—à–Ω–æ")
            else:
                print(f"‚ùå –û—à–∏–±–∫–∞ –ø–æ–ª—É—á–µ–Ω–∏—è —Ç–æ–∫–µ–Ω–∞: {response.status_code}")
        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ –ø–æ–ª—É—á–µ–Ω–∏—è —Ç–æ–∫–µ–Ω–∞: {e}")

    def extract_skills(self, text):
        """–ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –Ω–∞–≤—ã–∫–æ–≤ –∏–∑ —Ç–µ–∫—Å—Ç–∞ —Ä–µ–∑—é–º–µ"""
        prompt = f"""
        –ò–∑–≤–ª–µ–∫–∏ —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–∏–µ –Ω–∞–≤—ã–∫–∏ –∏ –∫–æ–º–ø–µ—Ç–µ–Ω—Ü–∏–∏ –∏–∑ —Ç–µ–∫—Å—Ç–∞ —Ä–µ–∑—é–º–µ.
        –í–µ—Ä–Ω–∏ –¢–û–õ–¨–ö–û JSON: {{"skills": ["skill1", "skill2", "skill3"]}}

        –¢–µ–∫—Å—Ç —Ä–µ–∑—é–º–µ: {text[:3000]}
        """

        messages = [
            {"role": "system", "content": "–¢—ã HR-—ç–∫—Å–ø–µ—Ä—Ç. –ò–∑–≤–ª–µ–∫–∞–µ—à—å –Ω–∞–≤—ã–∫–∏ –∏–∑ —Ä–µ–∑—é–º–µ. –û—Ç–≤–µ—á–∞–π —Ç–æ–ª—å–∫–æ JSON."},
            {"role": "user", "content": prompt}
        ]

        return self._get_chat_response(messages)

    def analyze_resume_match(self, resume_skills, vacancy_name, vacancy_skills):
        """–ê–Ω–∞–ª–∏–∑ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è —Ä–µ–∑—é–º–µ –≤–∞–∫–∞–Ω—Å–∏–∏"""
        prompt = f"""
        –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –≤–∞–∫–∞–Ω—Å–∏–∏ –∏ —Ä–∞—Å—Å—á–∏—Ç–∞–π –ø—Ä–æ—Ü–µ–Ω—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è —Ä–µ–∑—é–º–µ.

        –í–ê–ö–ê–ù–°–ò–Ø: {vacancy_name}
        –¢–†–ï–ë–£–ï–ú–´–ï –ù–ê–í–´–ö–ò: {', '.join(vacancy_skills)}

        –ù–ê–í–´–ö–ò –ö–ê–ù–î–ò–î–ê–¢–ê: {', '.join(resume_skills)}

        –í–µ—Ä–Ω–∏ –¢–û–õ–¨–ö–û JSON:
        {{
            "match_percentage": 85,
            "matched_skills": ["Python", "Django"],
            "missing_skills": ["Docker", "Kubernetes"],
            "recommendation": "–†–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞–Ω –∫ —Å–æ–±–µ—Å–µ–¥–æ–≤–∞–Ω–∏—é"
        }}
        """

        messages = [
            {"role": "system", "content": "–¢—ã Senior HR-–∞–Ω–∞–ª–∏—Ç–∏–∫. –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—à—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤."},
            {"role": "user", "content": prompt}
        ]

        return self._get_chat_response(messages)

    def _get_chat_response(self, messages):
        if not self.access_token:
            print("‚ùå –ù–µ—Ç access token")
            return None

        headers = {
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {self.access_token}',
        }

        payload = {
            'model': 'GigaChat',
            'messages': messages,
            'temperature': 0.3
        }

        try:
            response = requests.post(
                f"{self.config.GIGACHAT_API_URL}/chat/completions",
                headers=headers,
                json=payload,
                verify=False,
                timeout=30
            )

            if response.status_code == 200:
                return response.json()['choices'][0]['message']['content']
            else:
                print(f"‚ùå –û—à–∏–±–∫–∞ API: {response.status_code}")
        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ GigaChat: {e}")

        return None

print("‚úÖ GigaChat –∫–ª–∏–µ–Ω—Ç –≥–æ—Ç–æ–≤!")
```

```python
# üìä –ê–ù–ê–õ–ò–ó–ê–¢–û–† –†–ï–ó–Æ–ú–ï
class ResumeAnalyzer:
    def __init__(self):
        self.giga_client = GigaChatClient()
        self.config = Config()
        self.skill_model = SentenceTransformer('all-MiniLM-L6-v2')

    def analyze_resume_for_vacancies(self, resume_text):
        """–ê–Ω–∞–ª–∏–∑ —Ä–µ–∑—é–º–µ –¥–ª—è –≤—Å–µ—Ö –≤–∞–∫–∞–Ω—Å–∏–π"""
        print("üîç –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—é–º–µ...")

        # –ò–∑–≤–ª–µ–∫–∞–µ–º –Ω–∞–≤—ã–∫–∏ –∏–∑ —Ä–µ–∑—é–º–µ
        skills_response = self.giga_client.extract_skills(resume_text)
        resume_skills = self._parse_skills_from_response(skills_response)

        if not resume_skills:
            # –†–µ–∑–µ—Ä–≤–Ω—ã–π –º–µ—Ç–æ–¥ –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –Ω–∞–≤—ã–∫–æ–≤
            resume_skills = self._fallback_skill_extraction(resume_text)

        print(f"‚úÖ –ù–∞–π–¥–µ–Ω—ã –Ω–∞–≤—ã–∫–∏: {resume_skills}")

        # –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –∫–∞–∂–¥–æ–π –≤–∞–∫–∞–Ω—Å–∏–∏
        results = {}
        for vacancy_name, vacancy_data in self.config.VACANCIES.items():
            print(f"üìã –ü—Ä–æ–≤–µ—Ä—è–µ–º –≤–∞–∫–∞–Ω—Å–∏—é: {vacancy_name}")

            match_result = self.giga_client.analyze_resume_match(
                resume_skills,
                vacancy_name,
                vacancy_data["required_skills"]
            )

            # –ü–∞—Ä—Å–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
            vacancy_match = self._parse_match_result(match_result, vacancy_data)
            results[vacancy_name] = vacancy_match

            print(f"   üìä –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ: {vacancy_match['match_percentage']}%")

        return results, resume_skills

    def _parse_skills_from_response(self, response):
        """–ü–∞—Ä—Å–∏–Ω–≥ –Ω–∞–≤—ã–∫–æ–≤ –∏–∑ –æ—Ç–≤–µ—Ç–∞ GigaChat"""
        if not response:
            return []

        try:
            import re
            json_match = re.search(r'\{.*\}', response, re.DOTALL)
            if json_match:
                data = json.loads(json_match.group())
                return data.get('skills', [])
        except Exception as e:
            print(f"‚ö†Ô∏è –û—à–∏–±–∫–∞ –ø–∞—Ä—Å–∏–Ω–≥–∞ –Ω–∞–≤—ã–∫–æ–≤: {e}")

        return []

    def _parse_match_result(self, response, vacancy_data):
        """–ü–∞—Ä—Å–∏–Ω–≥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –∞–Ω–∞–ª–∏–∑–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è"""
        default_result = {
            "match_percentage": 0,
            "matched_skills": [],
            "missing_skills": vacancy_data["required_skills"],
            "recommendation": "–ù–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç",
            "vacancy_data": vacancy_data
        }

        if not response:
            return default_result

        try:
            import re
            json_match = re.search(r'\{.*\}', response, re.DOTALL)
            if json_match:
                data = json.loads(json_match.group())
                data["vacancy_data"] = vacancy_data
                return data
        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ –ø–∞—Ä—Å–∏–Ω–≥–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞: {e}")

        return default_result

    def _fallback_skill_extraction(self, text):
        """–†–µ–∑–µ—Ä–≤–Ω—ã–π –º–µ—Ç–æ–¥ –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –Ω–∞–≤—ã–∫–æ–≤"""
        common_skills = [
            "Python", "Java", "JavaScript", "SQL", "Linux", "Docker", "Kubernetes",
            "AWS", "Git", "React", "PostgreSQL", "MongoDB", "Django", "Flask",
            "Machine Learning", "Pandas", "NumPy", "HTML", "CSS", "TypeScript",
            "C++", "C#", "PHP", "Ruby", "Go", "Rust", "Swift", "Kotlin"
        ]

        found_skills = []
        text_lower = text.lower()

        for skill in common_skills:
            if skill.lower() in text_lower:
                found_skills.append(skill)

        return found_skills

print("‚úÖ –ê–Ω–∞–ª–∏–∑–∞—Ç–æ—Ä —Ä–µ–∑—é–º–µ –≥–æ—Ç–æ–≤!")
```

```python
# ü§ñ TELEGRAM –ë–û–¢ –° –û–¢–ü–†–ê–í–ö–û–ô –í –ö–ê–ù–ê–õ
from telegram import Update, ReplyKeyboardMarkup, KeyboardButton
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes

# –ò–º–ø–æ—Ä—Ç—ã –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ñ–∞–π–ª–∞–º–∏
try:
    import pdfplumber
    import docx
    HAS_DOCUMENT_SUPPORT = True
except ImportError as e:
    print(f"‚ö†Ô∏è –ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: {e}")
    HAS_DOCUMENT_SUPPORT = False

class HRTelegramBot:
    def __init__(self):
        self.config = Config()
        self.analyzer = ResumeAnalyzer()
        self.user_data = {}

    def get_main_keyboard(self):
        """–ö–ª–∞–≤–∏–∞—Ç—É—Ä–∞ –≥–ª–∞–≤–Ω–æ–≥–æ –º–µ–Ω—é"""
        keyboard = [
            [KeyboardButton("üìÑ –ó–∞–≥—Ä—É–∑–∏—Ç—å —Ä–µ–∑—é–º–µ"), KeyboardButton("üìù –°–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ")],
            [KeyboardButton("üîç –ù–∞–π—Ç–∏ –≤–∞–∫–∞–Ω—Å–∏–∏"), KeyboardButton("‚ÑπÔ∏è –ü–æ–º–æ—â—å")]
        ]
        return ReplyKeyboardMarkup(keyboard, resize_keyboard=True)

    async def send_to_channel(self, context: ContextTypes.DEFAULT_TYPE, message: str):
        """–§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –∫–∞–Ω–∞–ª"""
        try:
            await context.bot.send_message(
                chat_id=self.config.CHANNEL_ID,
                text=message,
                parse_mode='HTML'
            )
            print(f"‚úÖ –°–æ–æ–±—â–µ–Ω–∏–µ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–æ –≤ –∫–∞–Ω–∞–ª {self.config.CHANNEL_ID}")
            return True
        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ –æ—Ç–ø—Ä–∞–≤–∫–∏ –≤ –∫–∞–Ω–∞–ª: {e}")
            return False

    async def notify_new_user(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ –æ –Ω–æ–≤–æ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ –≤ –∫–∞–Ω–∞–ª"""
        user = update.effective_user

        welcome_message = f"""
<b>üÜï –ù–æ–≤—ã–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤ HR-–±–æ—Ç–µ!</b>

üë§ <b>–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è:</b>
–ò–º—è: {user.first_name}
–§–∞–º–∏–ª–∏—è: {user.last_name or '–Ω–µ —É–∫–∞–∑–∞–Ω–∞'}
Username: @{user.username or '–Ω–µ —É–∫–∞–∑–∞–Ω'}
ID: {user.id}

‚è∞ –í—Ä–µ–º—è: {update.message.date}
"""
        await self.send_to_channel(context, welcome_message)

    async def notify_resume_analysis(self, context: ContextTypes.DEFAULT_TYPE, user, results, skills):
        """–£–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ –æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞—Ö –∞–Ω–∞–ª–∏–∑–∞ —Ä–µ–∑—é–º–µ –≤ –∫–∞–Ω–∞–ª"""
        # –ù–∞—Ö–æ–¥–∏–º –ø–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏
        suitable_vacancies = {
            name: data for name, data in results.items()
            if data['match_percentage'] >= self.config.MATCH_THRESHOLD
        }

        if suitable_vacancies:
            message = f"""
<b>üéØ –ù–û–í–´–ô –ö–ê–ù–î–ò–î–ê–¢ –î–õ–Ø –í–ê–ö–ê–ù–°–ò–ô!</b>

üë§ <b>–ö–∞–Ω–¥–∏–¥–∞—Ç:</b>
–ò–º—è: {user.first_name}
Username: @{user.username or '–Ω–µ —É–∫–∞–∑–∞–Ω'}
ID: {user.id}

üõ†Ô∏è <b>–ù–∞–≤—ã–∫–∏ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞:</b>
{', '.join(skills[:10])}

‚úÖ <b>–ü–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏:</b>
"""
            for vacancy_name, data in suitable_vacancies.items():
                message += f"\nüè¢ <b>{vacancy_name}</b> - {data['match_percentage']}%"
                message += f"\nüíº {data['vacancy_data']['description']}"
                message += f"\nüí∞ {data['vacancy_data']['min_salary']:,} - {data['vacancy_data']['max_salary']:,} —Ä—É–±."
                if data['matched_skills']:
                    message += f"\n‚úÖ –°–æ–≤–ø–∞–≤—à–∏–µ –Ω–∞–≤—ã–∫–∏: {', '.join(data['matched_skills'][:3])}\n"

            message += f"\n‚è∞ –ê–Ω–∞–ª–∏–∑ –ø—Ä–æ–≤–µ–¥–µ–Ω: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}"

        else:
            # –õ—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–∞–∂–µ –µ—Å–ª–∏ –Ω–µ –ø–æ–¥—Ö–æ–¥–∏—Ç
            if results:
                best_match_name, best_match_data = max(results.items(), key=lambda x: x[1]['match_percentage'])
                message = f"""
<b>üìä –ö–ê–ù–î–ò–î–ê–¢ –ù–ï –ü–†–û–®–ï–õ –û–¢–ë–û–†</b>

üë§ <b>–ö–∞–Ω–¥–∏–¥–∞—Ç:</b>
–ò–º—è: {user.first_name}
Username: @{user.username or '–Ω–µ —É–∫–∞–∑–∞–Ω'}

üõ†Ô∏è <b>–ù–∞–≤—ã–∫–∏:</b>
{', '.join(skills[:8])}

üìà <b>–õ—É—á—à–µ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ:</b>
{best_match_name} - {best_match_data['match_percentage']}%

üí° <b>–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è:</b>
–¢—Ä–µ–±—É–µ—Ç—Å—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ

‚è∞ –í—Ä–µ–º—è –∞–Ω–∞–ª–∏–∑–∞: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
"""
            else:
                message = f"""
<b>üìä –û–®–ò–ë–ö–ê –ê–ù–ê–õ–ò–ó–ê</b>

üë§ <b>–ö–∞–Ω–¥–∏–¥–∞—Ç:</b>
–ò–º—è: {user.first_name}

‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—é–º–µ
"""

        await self.send_to_channel(context, message)

    async def start(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–ö–æ–º–∞–Ω–¥–∞ –Ω–∞—á–∞–ª–∞ —Ä–∞–±–æ—Ç—ã"""
        user = update.effective_user

        welcome_text = f"""
üëã –î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å, {user.first_name}!

–Ø - HR-–±–æ—Ç —Å –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–º –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç–æ–º. –Ø –ø–æ–º–æ–≥—É –≤–∞–º:

üìÑ –ê–Ω–∞–ª–∏–∑ —Ä–µ–∑—é–º–µ - –∑–∞–≥—Ä—É–∑–∏—Ç–µ –≤–∞—à–µ —Ä–µ–∑—é–º–µ
üìù –°–æ–∑–¥–∞–Ω–∏–µ —Ä–µ–∑—é–º–µ - –ø–æ–º–æ–≥—É —Å–æ—Å—Ç–∞–≤–∏—Ç—å –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω–æ–µ —Ä–µ–∑—é–º–µ
üîç –ü–æ–∏—Å–∫ –≤–∞–∫–∞–Ω—Å–∏–π - –Ω–∞–π–¥—É –ø–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏ –ø–æ –≤–∞—à–µ–º—É –æ–ø—ã—Ç—É

–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ –Ω–∏–∂–µ:
"""

        await update.message.reply_text(welcome_text, reply_markup=self.get_main_keyboard())

        # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º —É–≤–µ–¥–æ–º–ª–µ–Ω–∏–µ –≤ –∫–∞–Ω–∞–ª –æ –Ω–æ–≤–æ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ
        await self.notify_new_user(update, context)

        print(f"üë§ –ù–æ–≤—ã–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å: {user.first_name} (@{user.username})")

    async def handle_upload_resume(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ —Ä–µ–∑—é–º–µ"""
        if not HAS_DOCUMENT_SUPPORT:
            await update.message.reply_text(
                "‚ùå –ü–æ–¥–¥–µ—Ä–∂–∫–∞ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –≤—Ä–µ–º–µ–Ω–Ω–æ –Ω–µ–¥–æ—Å—Ç—É–ø–Ω–∞\n\n"
                "–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –æ–ø—Ü–∏—é 'üìù –°–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ' –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Ä–µ–∑—é–º–µ –≤ —á–∞—Ç–µ."
            )
            return

        instruction = """
üìÑ –ó–∞–≥—Ä—É–∑–∫–∞ —Ä–µ–∑—é–º–µ

–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –æ—Ç–ø—Ä–∞–≤—å—Ç–µ –≤–∞—à–µ —Ä–µ–∑—é–º–µ –≤ –æ–¥–Ω–æ–º –∏–∑ —Ñ–æ—Ä–º–∞—Ç–æ–≤:
‚Ä¢ PDF
‚Ä¢ DOC/DOCX
‚Ä¢ TXT

–Ø –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É—é –µ–≥–æ –∏ –Ω–∞–π–¥—É –ø–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏!
"""
        await update.message.reply_text(instruction)

    async def handle_create_resume(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–°–æ–∑–¥–∞–Ω–∏–µ —Ä–µ–∑—é–º–µ —á–µ—Ä–µ–∑ –¥–∏–∞–ª–æ–≥"""
        questions = [
            "–ö–∞–∫ –≤–∞—Å –∑–æ–≤—É—Ç (–§–ò–û)?",
            "–ö–∞–∫–∞—è —É –≤–∞—Å –ø—Ä–æ—Ñ–µ—Å—Å–∏—è/—Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ—Å—Ç—å?",
            "–°–∫–æ–ª—å–∫–æ –ª–µ—Ç –æ–ø—ã—Ç–∞ —Ä–∞–±–æ—Ç—ã?",
            "–ö–∞–∫–∏–µ –∫–ª—é—á–µ–≤—ã–µ –Ω–∞–≤—ã–∫–∏ –∏ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏ –≤—ã –∑–Ω–∞–µ—Ç–µ?",
            "–í–∞—à —É—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è?",
            "–ö–∞–∫–∏–µ –ø—Ä–æ–µ–∫—Ç—ã –∏–ª–∏ –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è —Ö–æ—Ç–∏—Ç–µ –æ—Ç–º–µ—Ç–∏—Ç—å?",
            "–í–∞—à–∏ –∫–æ–Ω—Ç–∞–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ (email/—Ç–µ–ª–µ—Ñ–æ–Ω)?"
        ]

        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤–æ–ø—Ä–æ—Å—ã –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        user_id = update.effective_user.id
        self.user_data[user_id] = {
            'resume_questions': questions,
            'current_question': 0,
            'resume_answers': [],
            'resume_text': ""
        }

        await update.message.reply_text(
            "üìù –°–æ–∑–¥–∞–Ω–∏–µ —Ä–µ–∑—é–º–µ\n\n"
            "–û—Ç–ª–∏—á–Ω–æ! –î–∞–≤–∞–π—Ç–µ —Å–æ–∑–¥–∞–¥–∏–º –≤–∞—à–µ –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω–æ–µ —Ä–µ–∑—é–º–µ.\n"
            "–Ø –∑–∞–¥–∞–º –≤–∞–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –≤–æ–ø—Ä–æ—Å–æ–≤.\n\n"
            f"‚ùì –í–æ–ø—Ä–æ—Å 1/{len(questions)}:\n{questions[0]}"
        )

    async def handle_find_vacancies(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–ü–æ–∏—Å–∫ –≤–∞–∫–∞–Ω—Å–∏–π –±–µ–∑ —Ä–µ–∑—é–º–µ"""
        instruction = """
üîç –ü–æ–∏—Å–∫ –≤–∞–∫–∞–Ω—Å–∏–π

–î–ª—è –ø–æ–∏—Å–∫–∞ –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö –≤–∞–∫–∞–Ω—Å–∏–π –º–Ω–µ –Ω—É–∂–Ω–æ –≤–∞—à–µ —Ä–µ–∑—é–º–µ.

–í—ã –º–æ–∂–µ—Ç–µ:
1. üìÑ –ó–∞–≥—Ä—É–∑–∏—Ç—å –≥–æ—Ç–æ–≤–æ–µ —Ä–µ–∑—é–º–µ
2. üìù –°–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ –ø—Ä—è–º–æ –∑–¥–µ—Å—å

–ü–æ—Å–ª–µ –∞–Ω–∞–ª–∏–∑–∞ —è –ø–æ–∫–∞–∂—É –≤–∞–∫–∞–Ω—Å–∏–∏, –≥–¥–µ –≤–∞—à–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –±–æ–ª–µ–µ 70%!
"""
        await update.message.reply_text(instruction, reply_markup=self.get_main_keyboard())

    async def handle_document(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤"""
        if not HAS_DOCUMENT_SUPPORT:
            await update.message.reply_text(
                "‚ùå –ü–æ–¥–¥–µ—Ä–∂–∫–∞ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –≤—Ä–µ–º–µ–Ω–Ω–æ –Ω–µ–¥–æ—Å—Ç—É–ø–Ω–∞\n\n"
                "–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –æ–ø—Ü–∏—é 'üìù –°–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ' –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Ä–µ–∑—é–º–µ –≤ —á–∞—Ç–µ."
            )
            return

        document = update.message.document
        user = update.effective_user

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ñ–æ—Ä–º–∞—Ç —Ñ–∞–π–ª–∞
        allowed_extensions = ['.pdf', '.doc', '.docx', '.txt']
        file_extension = f".{document.file_name.split('.')[-1].lower()}"

        if file_extension not in allowed_extensions:
            await update.message.reply_text(
                "‚ùå –ù–µ–≤–µ—Ä–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç —Ñ–∞–π–ª–∞!\n\n"
                "–ü–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–µ —Ñ–æ—Ä–º–∞—Ç—ã: PDF, DOC, DOCX, TXT\n"
                f"–í–∞—à —Ñ–∞–π–ª: {document.file_name}"
            )
            return

        # –£–≤–µ–¥–æ–º–ª—è–µ–º –æ –Ω–∞—á–∞–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏
        processing_msg = await update.message.reply_text(
            "üîÑ –û–±—Ä–∞–±–∞—Ç—ã–≤–∞—é –≤–∞—à–µ —Ä–µ–∑—é–º–µ...\n\n"
            "–ê–Ω–∞–ª–∏–∑–∏—Ä—É—é –Ω–∞–≤—ã–∫–∏ –∏ –∏—â—É –ø–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏.\n"
            "–≠—Ç–æ –∑–∞–π–º–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–µ–∫—É–Ω–¥..."
        )

        try:
            # –°–∫–∞—á–∏–≤–∞–µ–º —Ñ–∞–π–ª
            file = await document.get_file()
            file_path = f"temp_{user.id}_{document.file_name}"
            await file.download_to_drive(file_path)

            # –ò–∑–≤–ª–µ–∫–∞–µ–º —Ç–µ–∫—Å—Ç
            resume_text = self.extract_text_from_file(file_path)

            if not resume_text:
                await update.message.reply_text(
                    "‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –ø—Ä–æ—á–∏—Ç–∞—Ç—å —Ñ–∞–π–ª\n\n"
                    "–ü–æ–ø—Ä–æ–±—É–π—Ç–µ –¥—Ä—É–≥–æ–π —Ñ–æ—Ä–º–∞—Ç –∏–ª–∏ —Å–æ–∑–¥–∞–π—Ç–µ —Ä–µ–∑—é–º–µ –≤ —á–∞—Ç–µ."
                )
                return

            # –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—é–º–µ
            results, skills = self.analyzer.analyze_resume_for_vacancies(resume_text)

            # –§–æ—Ä–º–∏—Ä—É–µ–º –∏ –æ—Ç–ø—Ä–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
            await self.send_analysis_results(update, context, results, skills, user)

            # –£–¥–∞–ª—è–µ–º –≤—Ä–µ–º–µ–Ω–Ω—ã–π —Ñ–∞–π–ª
            import os
            if os.path.exists(file_path):
                os.remove(file_path)

        except Exception as e:
            await update.message.reply_text(
                f"‚ùå –û—à–∏–±–∫–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏: {str(e)}\n\n"
                "–ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑ –∏–ª–∏ —Å–æ–∑–¥–∞–π—Ç–µ —Ä–µ–∑—é–º–µ –≤ —á–∞—Ç–µ."
            )

    def extract_text_from_file(self, file_path):
        """–ò–∑–≤–ª–µ—á–µ–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ –∏–∑ —Ñ–∞–π–ª–∞"""
        text = ""
        try:
            if file_path.endswith('.pdf'):
                with pdfplumber.open(file_path) as pdf:
                    for page in pdf.pages:
                        if page.extract_text():
                            text += page.extract_text() + "\n"
            elif file_path.endswith('.docx'):
                doc = docx.Document(file_path)
                text = "\n".join([paragraph.text for paragraph in doc.paragraphs])
            elif file_path.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8') as f:
                    text = f.read()
        except Exception as e:
            print(f"‚ùå –û—à–∏–±–∫–∞ —á—Ç–µ–Ω–∏—è —Ñ–∞–π–ª–∞: {e}")
        return text

    async def send_analysis_results(self, update: Update, context: ContextTypes.DEFAULT_TYPE, results, skills, user):
        """–û—Ç–ø—Ä–∞–≤–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∞–Ω–∞–ª–∏–∑–∞"""
        # –ù–∞—Ö–æ–¥–∏–º –ø–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏ (match > 70%)
        suitable_vacancies = {
            name: data for name, data in results.items()
            if data['match_percentage'] >= self.config.MATCH_THRESHOLD
        }

        # –§–æ—Ä–º–∏—Ä—É–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        if suitable_vacancies:
            message = "üéâ –û—Ç–ª–∏—á–Ω—ã–µ –Ω–æ–≤–æ—Å—Ç–∏!\n\n"
            message += "‚úÖ –í—ã –ø–æ–¥—Ö–æ–¥–∏—Ç–µ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏:\n\n"

            for vacancy_name, data in suitable_vacancies.items():
                message += f"üè¢ {vacancy_name}\n"
                message += f"üìä –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ: {data['match_percentage']}%\n"
                message += f"üíº {data['vacancy_data']['description']}\n"
                message += f"üí∞ –ó–∞—Ä–ø–ª–∞—Ç–∞: {data['vacancy_data']['min_salary']:,} - {data['vacancy_data']['max_salary']:,} —Ä—É–±.\n"

                if data['matched_skills']:
                    message += f"‚úÖ –í–∞—à–∏ –Ω–∞–≤—ã–∫–∏: {', '.join(data['matched_skills'][:5])}\n"

                message += "\n"

            message += "üìû –ú–µ–Ω–µ–¥–∂–µ—Ä –∫–æ–º–ø–∞–Ω–∏–∏ —Å–≤—è–∂–µ—Ç—Å—è —Å –≤–∞–º–∏ –≤ –±–ª–∏–∂–∞–π—à–µ–µ –≤—Ä–µ–º—è!\n\n"
            message += "–•–æ—Ç–∏—Ç–µ –Ω–∞–π—Ç–∏ –¥—Ä—É–≥–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏? –ó–∞–≥—Ä—É–∑–∏—Ç–µ –æ–±–Ω–æ–≤–ª–µ–Ω–Ω–æ–µ —Ä–µ–∑—é–º–µ!"

        else:
            message = "üìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–Ω–∞–ª–∏–∑–∞:\n\n"
            message += "–ö —Å–æ–∂–∞–ª–µ–Ω–∏—é, –ø–æ —Ç–µ–∫—É—â–µ–º—É —Ä–µ–∑—é–º–µ –Ω–µ—Ç –≤–∞–∫–∞–Ω—Å–∏–π —Å –≤—ã—Å–æ–∫–∏–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ–º.\n\n"

            # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ª—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç
            if results:
                best_match = max(results.items(), key=lambda x: x[1]['match_percentage'])
                message += f"üß≠ –°–∞–º—ã–π –±–ª–∏–∑–∫–∏–π –≤–∞—Ä–∏–∞–Ω—Ç: {best_match[0]}\n"
                message += f"üìä –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ: {best_match[1]['match_percentage']}%\n\n"

            message += "üí° –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:\n"
            message += "‚Ä¢ –û–±–Ω–æ–≤–∏—Ç–µ —Ä–µ–∑—é–º–µ, –¥–æ–±–∞–≤–∏–≤ –±–æ–ª—å—à–µ –Ω–∞–≤—ã–∫–æ–≤\n"
            message += "‚Ä¢ –†–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ —Å–º–µ–∂–Ω—ã–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è\n"
            message += "‚Ä¢ –ü—Ä–æ–π–¥–∏—Ç–µ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∫—É—Ä—Å—ã\n\n"
            message += "–ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ –∑–∞–Ω–æ–≤–æ —Å –º–æ–µ–π –ø–æ–º–æ—â—å—é!"

        await update.message.reply_text(message, reply_markup=self.get_main_keyboard())

        # üî• –û–¢–ü–†–ê–í–õ–Ø–ï–ú –£–í–ï–î–û–ú–õ–ï–ù–ò–ï –í –ö–ê–ù–ê–õ
        try:
            await self.notify_resume_analysis(context, user, results, skills)
        except Exception as e:
            print(f"‚ö†Ô∏è –ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–ø—Ä–∞–≤–∏—Ç—å –≤ –∫–∞–Ω–∞–ª: {e}")

        # –õ–æ–≥–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
        print(f"üìä –ê–Ω–∞–ª–∏–∑ –¥–ª—è {user.first_name}: {len(suitable_vacancies)} –ø–æ–¥—Ö–æ–¥—è—â–∏—Ö –≤–∞–∫–∞–Ω—Å–∏–π")

    async def handle_text_message(self, update: Update, context: ContextTypes.DEFAULT_TYPE):
        """–û–±—Ä–∞–±–æ—Ç–∫–∞ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π"""
        text = update.message.text
        user_id = update.effective_user.id

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å–æ–∑–¥–∞–Ω–∏—è —Ä–µ–∑—é–º–µ
        if user_id in self.user_data and 'resume_questions' in self.user_data[user_id]:
            await self.handle_resume_creation(update, user_id, context)
            return

        # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–æ–º–∞–Ω–¥ –∏–∑ –∫–Ω–æ–ø–æ–∫
        if text == "üìÑ –ó–∞–≥—Ä—É–∑–∏—Ç—å —Ä–µ–∑—é–º–µ":
            await self.handle_upload_resume(update, context)
        elif text == "üìù –°–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ":
            await self.handle_create_resume(update, context)
        elif text == "üîç –ù–∞–π—Ç–∏ –≤–∞–∫–∞–Ω—Å–∏–∏":
            await self.handle_find_vacancies(update, context)
        elif text == "‚ÑπÔ∏è –ü–æ–º–æ—â—å":
            await self.send_help(update)
        else:
            await update.message.reply_text(
                "–Ø –Ω–µ –ø–æ–Ω—è–ª –≤–∞—à –∑–∞–ø—Ä–æ—Å. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–Ω–æ–ø–∫–∏ –Ω–∏–∂–µ –¥–ª—è –Ω–∞–≤–∏–≥–∞—Ü–∏–∏.",
                reply_markup=self.get_main_keyboard()
            )

    async def handle_resume_creation(self, update: Update, user_id, context: ContextTypes.DEFAULT_TYPE):
        """–û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ—Ü–µ—Å—Å–∞ —Å–æ–∑–¥–∞–Ω–∏—è —Ä–µ–∑—é–º–µ"""
        user_data = self.user_data[user_id]
        questions = user_data['resume_questions']
        current_index = user_data['current_question']
        answers = user_data['resume_answers']

        # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ç–µ–∫—É—â–∏–π –æ—Ç–≤–µ—Ç
        answers.append(update.message.text)

        # –ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É –≤–æ–ø—Ä–æ—Å—É
        next_index = current_index + 1

        if next_index < len(questions):
            user_data['current_question'] = next_index
            await update.message.reply_text(
                f"‚úÖ –û—Ç–≤–µ—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω!\n\n"
                f"‚ùì –í–æ–ø—Ä–æ—Å {next_index + 1}/{len(questions)}:\n{questions[next_index]}"
            )
        else:
            # –í—Å–µ –≤–æ–ø—Ä–æ—Å—ã –æ—Ç–≤–µ—á–µ–Ω—ã, —Å–æ–∑–¥–∞–µ–º —Ä–µ–∑—é–º–µ
            await self.create_final_resume(update, user_id, answers, context)

    async def create_final_resume(self, update: Update, user_id, answers, context: ContextTypes.DEFAULT_TYPE):
        """–°–æ–∑–¥–∞–Ω–∏–µ —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —Ä–µ–∑—é–º–µ –∏ –∞–Ω–∞–ª–∏–∑"""
        # –§–æ—Ä–º–∏—Ä—É–µ–º —Ç–µ–∫—Å—Ç —Ä–µ–∑—é–º–µ
        resume_text = self.format_resume_text(answers)

        # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ç–µ–∫—Å—Ç —Ä–µ–∑—é–º–µ
        self.user_data[user_id]['resume_text'] = resume_text

        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Ä–µ–∑—é–º–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é
        await update.message.reply_text(
            "üìã –í–∞—à–µ —Ä–µ–∑—é–º–µ –≥–æ—Ç–æ–≤–æ!\n\n"
            f"{resume_text}\n\n"
            "üîÑ –ê–Ω–∞–ª–∏–∑–∏—Ä—É—é –≤–∞–∫–∞–Ω—Å–∏–∏..."
        )

        # –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—é–º–µ
        results, skills = self.analyzer.analyze_resume_for_vacancies(resume_text)

        # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
        await self.send_analysis_results(update, context, results, skills, update.effective_user)

        # –û—á–∏—â–∞–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        del self.user_data[user_id]

    def format_resume_text(self, answers):
        """–§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ —Ä–µ–∑—é–º–µ –∏–∑ –æ—Ç–≤–µ—Ç–æ–≤"""
        name, profession, experience, skills, education, projects, contacts = answers

        resume_text = f"""
üë§ –§–ò–û: {name}
üíº –ü—Ä–æ—Ñ–µ—Å—Å–∏—è: {profession}
üìÖ –û–ø—ã—Ç —Ä–∞–±–æ—Ç—ã: {experience}
üõ†Ô∏è –ù–∞–≤—ã–∫–∏: {skills}
üéì –û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ: {education}
üöÄ –ü—Ä–æ–µ–∫—Ç—ã –∏ –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è: {projects}
üìû –ö–æ–Ω—Ç–∞–∫—Ç—ã: {contacts}
"""
        return resume_text

    async def send_help(self, update: Update):
        """–û—Ç–ø—Ä–∞–≤–∫–∞ —Å–ø—Ä–∞–≤–∫–∏"""
        help_text = """
‚ÑπÔ∏è –ü–æ–º–æ—â—å –ø–æ –±–æ—Ç—É

ü§ñ –ß—Ç–æ —è —É–º–µ—é:
‚Ä¢ –ê–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –≤–∞—à–µ —Ä–µ–∑—é–º–µ —Å –ø–æ–º–æ—â—å—é –ò–ò
‚Ä¢ –ù–∞—Ö–æ–¥–∏—Ç—å –ø–æ–¥—Ö–æ–¥—è—â–∏–µ –≤–∞–∫–∞–Ω—Å–∏–∏
‚Ä¢ –ü–æ–º–æ–≥–∞—Ç—å —Å–æ–∑–¥–∞–≤–∞—Ç—å –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω—ã–µ —Ä–µ–∑—é–º–µ
‚Ä¢ –û—Ü–µ–Ω–∏–≤–∞—Ç—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º —Ä–∞–±–æ—Ç–æ–¥–∞—Ç–µ–ª–µ–π

üìã –ö–∞–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è:
1. üìÑ –ó–∞–≥—Ä—É–∑–∏—Ç—å —Ä–µ–∑—é–º–µ - –æ—Ç–ø—Ä–∞–≤—å—Ç–µ —Ñ–∞–π–ª (PDF, DOC, DOCX, TXT)
2. üìù –°–æ–∑–¥–∞—Ç—å —Ä–µ–∑—é–º–µ - –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã –¥–ª—è —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Ä–µ–∑—é–º–µ
3. üîç –ù–∞–π—Ç–∏ –≤–∞–∫–∞–Ω—Å–∏–∏ - —É–∑–Ω–∞–π—Ç–µ, –∫—É–¥–∞ –≤—ã –ø–æ–¥—Ö–æ–¥–∏—Ç–µ

üéØ –ö—Ä–∏—Ç–µ—Ä–∏–∏ –æ—Ç–±–æ—Ä–∞:
‚Ä¢ –í–∞–∫–∞–Ω—Å–∏–∏ —Å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ–º >70% —Å—á–∏—Ç–∞—é—Ç—Å—è –ø–æ–¥—Ö–æ–¥—è—â–∏–º–∏
‚Ä¢ –ú–µ–Ω–µ–¥–∂–µ—Ä —Å–≤—è–∂–µ—Ç—Å—è —Å –≤–∞–º–∏ –ø–æ –ø–æ–¥—Ö–æ–¥—è—â–∏–º –≤–∞–∫–∞–Ω—Å–∏—è–º

–ü–æ –≤—Å–µ–º –≤–æ–ø—Ä–æ—Å–∞–º: @hr_manager
"""
        await update.message.reply_text(help_text)

    def run_bot(self):
        """–ó–∞–ø—É—Å–∫ –±–æ—Ç–∞"""
        app = Application.builder().token(self.config.BOT_TOKEN).build()

        # –û–±—Ä–∞–±–æ—Ç—á–∏–∫–∏ –∫–æ–º–∞–Ω–¥
        app.add_handler(CommandHandler("start", self.start))
        app.add_handler(CommandHandler("help", self.send_help))

        # –û–±—Ä–∞–±–æ—Ç—á–∏–∫–∏ —Å–æ–æ–±—â–µ–Ω–∏–π
        app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, self.handle_text_message))

        if HAS_DOCUMENT_SUPPORT:
            app.add_handler(MessageHandler(filters.Document.ALL, self.handle_document))

        print("ü§ñ HR Bot –∑–∞–ø—É—â–µ–Ω...")
        print("üîÑ –û–∂–∏–¥–∞–µ–º —Å–æ–æ–±—â–µ–Ω–∏—è...")

        app.run_polling()

print("‚úÖ Telegram –±–æ—Ç –≥–æ—Ç–æ–≤!")
```
